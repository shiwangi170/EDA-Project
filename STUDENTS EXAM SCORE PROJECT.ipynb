## DROP UNNAMED COLUMN

```python
df=df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("5 - 10","5 to 10")  #for changing value of wklyStudyHours
df.head()
```

## gender distribution

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("GENDER DISTRIBUTION")
plt.show()
```

From the above chart we have analysed that 
The number of females in the data is more than the number of males

## CREATE GROUPBY

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gb
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot= True)
plt.title("Relationship between Parent's education and student's score")
plt.show()
```

From the above chart we have concluded that the education of the parents have a good impact

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gb1
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot= True)
plt.title("Relationship between Parent's marital status and Student's score")
plt.show()
```

From the above chart we have concluded that the marital status of the parents have no impact or negligible impact on the student's score

## DISTRIBUTION OF Ethnic Group

```python
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()

l=["groupA","groupB","groupC","groupD","groupE"]
mylist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mylist,labels=l,autopct="%1.2f%%")
plt.title("Distributions of ethnic groups")
plt.show()
```
