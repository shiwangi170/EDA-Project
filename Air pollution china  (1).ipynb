```python
df.describe(include='all').plot(kind='bar', figsize=(6,4))
plt.title('Summary Statistics of DataFrame')
plt.ylabel('Values')
plt.xlabel('Features')
plt.show()
```

df.describe(include='all') will generate summary statistics for all columns in the DataFrame, including non-numeric ones 
and
.plot(kind='bar') will plot the results of the describe() method as a bar chart.

```python
plt.figure(figsize=(6,3))
sns.lineplot(data=df, x='Year', y='PM2.5 (µg/m³)', hue='City',errorbar=None)
plt.title('PM2.5 Trend by Year and City')
plt.ylabel('PM2.5 (µg/m³)')
plt.show()
```

```python
df.sort_index()[['Temperature (°C)','Humidity (%)']]
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(data=df, x='Season', y='PM2.5 (µg/m³)',color='g')
plt.title('PM2.5 Distribution by Season')
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x='Season',data=df,color='y')
plt.title('The season wise count')
plt.show()
```

```python
category_col = df.select_dtypes(include='object').columns 

plt.figure(figsize=(3,2))
for col in category_col:
    sns.countplot(x=col,data=df,color='yellow')
    plt.title(f'The distribution of {col}')
    plt.ylabel('Frequency')
    plt.show()
```

```python
numerical_col = df.select_dtypes(include=np.number).columns 
```

```python
plt.figure(figsize=(4,3))
for col in numerical_col:
    sns.histplot(x=col,data=df,kde=True,color='pink')
    plt.title(f'The distribution of {col}')
    plt.ylabel('Frequency')
    plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(df[numerical_col].corr(),annot=True,cmap='plasma')
plt.title('The heatmap data info')
plt.show()
```

```python
df.columns.to_list
```

```python
plt.figure(figsize=(6,3))
sns.boxplot(x='City',y='Humidity (%)',data=df,color="red")
plt.title('The city wise humidity')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Year',y='Wind Speed (m/s)',data=df,color='b',marker='*')
plt.title('The year wise wind speed')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.lineplot(x='Month',y='Hour',data=df,color='orange',marker='*')
plt.title('The hour wise month')
plt.show()
```

```python
pivot = df.pivot_table(index='Day of Week', columns='Hour', values='AQI', aggfunc='mean')
plt.figure(figsize=(6,4))
sns.heatmap(pivot, cmap='YlOrRd')
plt.title('Average AQI by Day of Week and Hour')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Temperature (°C)', y='PM2.5 (µg/m³)', hue='Season', alpha=0.6)
plt.title('PM2.5 vs Temperature by Season')
plt.grid(True)
plt.show()
```

```python
weather_aqi = df.groupby('Weather Condition')['AQI'].mean().sort_values(ascending=False)
plt.figure(figsize=(6,4))
sns.barplot(x=weather_aqi.values, y=weather_aqi.index)
plt.title('Average AQI by Weather Condition')
plt.xlabel('AQI')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Longitude', y='Latitude', hue='AQI', size='PM2.5 (µg/m³)',palette='coolwarm', alpha=0.5, sizes=(5, 200))
plt.title('Geospatial Distribution of AQI and PM2.5')
plt.legend(bbox_to_anchor=(1.01, 1), loc=2)
plt.show()
```
